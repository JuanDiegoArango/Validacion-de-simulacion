```python
lp=1.2 #nm
Kb_T=4.10 # pNâ‹…nm
#velocidad de elongacion 1 um/s.
```

Funciones

```python
def fuerza(x,c):
    return (Kb_T/(lp))*(0.25*(1.0-(x/c))**-2.0-0.25+(x/c))
def probabilidad_de_transicion(k,f,delta_X,dt):
    return k*exp((delta_X*f)/(Kb_T))*dt
```

```python
def flutuaciones(cambio_L,dt):
    ka=(3.36)**2*(41) #pN/nm
    kAB=5.0e-2 #Hz
    kBA=7.0e2 #Hz
    delta_X_AB=0.5 #nm
    delta_X_BA=-0.5 #nm
    Ntot=250.0
    Na=Ntot-5
    Nb=5
    La0=0.1 #nm
    Lb0=5.0 #nm
    L_inicial=Na*La0
    L=Na*La0
    cp=zeros(4)
    fuerzas=[]
    j=0
    while (L-L_inicial<=1000.0):  
        a=La0*Na
        b=Kb_T/(lp)
        c=Lb0*Nb
        L=L_inicial+cambio_L*dt*(j)
        if (Na!=0):
            cp[0]=4*(Na*b+c*ka)
            cp[1]=-(c)*(8*c*ka+4*ka*(L-a)+9*Na*b)
            cp[2]=(2*c**2)*(2*c*ka+4*ka*(L-a)+3*Na*b)
            cp[3]=4*(c**3)*(ka)*(a-L)
            raices=roots(cp)
            Lb=raices[2]
            f=(fuerza(Lb,c))
            if(L-L_inicial>=200):
                fuerzas.append(f)
      
        p=random.uniform(size=2)
        pb=probabilidad_de_transicion(kAB,(f),delta_X_AB,dt)
        pa=probabilidad_de_transicion(kBA,(f),delta_X_BA,dt)  
        
        if (pa>p[0]) & (Na<Ntot-5) & (Nb<5) & (Na+Nb==Ntot): 
            Na=Na+1.0
            Nb=Nb-1.0
            
        if (pb>p[1]) & (Nb<Ntot) & (Na>0) & (Na+Nb==Ntot):
            Na=Na-1.0
            Nb=Nb+1.0
        j=j+1

    return [mean((fuerzas)),std(fuerzas)]
```

```python
a=[1,2,3,4,5,6]
b=[9,8,7,6,5,4,3,2,1]
dt=[]
promedio01=[]
ruido01=[]
for i in a:
    print i
    for j in (b):
        dt.append(j*10**(-i))
        c=flutuaciones(0.1e3,(j*10**(-i)))     
        promedio01.append(c[0])
        ruido01.append(c[1])
```

```python
a=[1,2,3,4,5,6]
b=[9,8,7,6,5,4,3,2,1]
dt=[]
promedio1=[]
ruido1=[]
for i in a:
    print i
    for j in (b):
        dt.append(j*10**(-i))
        c=flutuaciones(1.0e3,(j*10**(-i)))     
        promedio1.append(c[0])
        ruido1.append(c[1])
```

```python
a=[1,2,3,4,5,6]
b=[9,8,7,6,5,4,3,2,1]
dt=[]
promedio10=[]
ruido10=[]
for i in a:
    print i
    for j in (b):
        dt.append(j*10**(-i))
        c=flutuaciones(10.0e3,(j*10**(-i)))     
        promedio10.append(c[0])
        ruido10.append(c[1])
```

```python
pylab.rcParams['figure.figsize'] = (24, 16)
plt.figure(1)

plt.subplot(221)
loglog(dt,promedio01,label="Mean")
loglog(dt,ruido01,label="Std. Deviation")
xlabel("dt(s)")
title("Extension velocity: 0.1 um/s")
plt.legend(bbox_to_anchor=(0.0, 1), loc=2, borderaxespad=0.)
ylim(min(ruido01),max(promedio01))

plt.subplot(222)
loglog(dt,promedio1,label="Mean")
loglog(dt,ruido1,label="Std. Deviation")
xlabel("dt(s)")
title("Extension velocity: 1.0 um/s")
plt.legend(bbox_to_anchor=(0.0, 1), loc=2, borderaxespad=0.)
ylim(min(ruido1),max(promedio1))


plt.subplot(223)
loglog(dt,promedio10,label="Mean")
loglog(dt,ruido10,label="Std. Deviation")
xlabel("dt(s)")
title("Extension velocity: 10.0 um/s")
plt.legend(bbox_to_anchor=(0.0, 1), loc=2, borderaxespad=0.)
ylim(min(ruido10),max(promedio10))


plt.subplot(224)
title("Comparison of Std.Deviation")

loglog(dt,ruido01,label="Std. for 0.1 um/s")
loglog(dt,ruido1,label="Std. for 1.0 um/s")
loglog(dt,ruido10,label="Std. for 10.0 um/s")
xlabel("dt(s)")


plt.legend(bbox_to_anchor=(0.0, 1), loc=2, borderaxespad=0.)
savefig('Fluctuaciones_Pili.png')

```
